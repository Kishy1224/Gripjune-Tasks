```python
# Visualizing the correlation between the Attributes
sns.heatmap(df.corr(), annot=True)
print("1 represents strong positive correlation")
print("-0.22 represents negative correlation")
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df.Sales,shade=True,color='red')
sns.kdeplot(df.Profit,shade=True,color='blue')
plt.xlim(-100,1000)
```

```python
# Countplot
fig,axs=plt.subplots(nrows=2,ncols=2,figsize=(10,7));

sns.countplot(df['Category'],ax=axs[0][0])
axs[0][0].set_title('Category',fontsize=20)

sns.countplot(df['Segment'],ax=axs[0][1])
axs[0][1].set_title('Segment',fontsize=20)

sns.countplot(df['Ship Mode'],ax=axs[1][0])
axs[1][0].set_title('Ship Mode',fontsize=20)

sns.countplot(df['Region'],ax=axs[1][1])
axs[1][1].set_title('Region',fontsize=20)
plt.tight_layout()
```

```python
# Countplot State wise shipments
plt.figure(figsize=(12,7))
sns.countplot(x=df['State'])
plt.xticks(rotation=-90)
plt.title('Count of State wise shipments')
```

## California has highest sales. -Some states like Montana, Vermont, Maine, Wyoming and West Virginia have less or no sales

```python
#state vs profit
plt.figure(figsize = (20,12))
df.groupby(by = 'State')['Profit'].sum().sort_values(ascending = True).plot(kind = 'bar')
```

```python
# Count of Sub-Category
plt.figure(figsize=(12,7))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.title('Count of Sub-categories')
```

```python
# Discount Vs Profit
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
```

```python
# Discount VS Sales
sns.lineplot(x='Discount',y='Sales',label='Profit',data=df)
plt.legend()
```

```python
# Count of Ship Mode by Region
plt.figure(figsize=(9,5))
sns.countplot(x='Region',hue='Ship Mode',data=df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of Ship Mode by Region')
```
